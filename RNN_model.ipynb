```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# dataset
texts = [
    "i love this movie",
    "this film is great",
    "i really like this",
    "this is amazing",
    "i hate this movie",
    "this film is bad",
    "i do not like this",
    "this is terrible"
]

labels = np.array([1, 1, 1, 1, 0, 0, 0, 0])  # 1 = positive, 0 = negative

# text preprocessing
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

max_len = max(len(seq) for seq in sequences)
X = pad_sequences(sequences, maxlen=max_len, padding='post')

# build RNN model
vocab_size = len(tokenizer.word_index) + 1

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=8),
    SimpleRNN(16),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# training
model.fit(X, labels, epochs=20, verbose=1)

# test examples
test_texts = ["i love this film", "this movie is bad"]
test_seq = tokenizer.texts_to_sequences(test_texts)
test_pad = pad_sequences(test_seq, maxlen=max_len, padding='post')

preds = model.predict(test_pad)

for t, p in zip(test_texts, preds):
    print(t, "->", "Positive" if p > 0.5 else "Negative")
```
